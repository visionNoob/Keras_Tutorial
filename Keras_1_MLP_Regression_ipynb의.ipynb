# Linear Regression (선형 회귀) 예제

# 1. 라이브러리 불러오기

```python
# 1.라이브러리 불러오기
import tensorflow as tf # 딥러닝 라이브러리
import keras # 딥러닝 라이브러리
import numpy as np # 수학 계산 라이브러리
import matplotlib.pyplot as plt # 그래프 그리는 라이브러리 

#항상 버전 확인은 필수
print(tf.__version__)
print(keras.__version__)
```

# 2. 랜덤 시드 설정하기 (중요!)
https://keras.io/getting-started/faq/#how-can-i-obtain-reproducible-results-using-keras-during-development

```python
# 2.랜덤 시드 고정하기
import random
seed_value = 42 # 랜덤 시드

random.seed(seed_value)
np.random.seed(seed_value)
tf.set_random_seed(seed_value)
```

# 3. 데이터셋

```python
# 3. 데이터셋
# 입력 데이터
input_data = np.array([1, 2, 3, 4, 5])
output_data = np.array([0, 3, 2, 5, 4]) 

input_data = input_data - 3
output_data = output_data - 3

# 검증 데이터
input_val_data = np.array([1.5, 2.5, 3.5, 4.5, 5.5])
output_val_data = np.array([1.5, 2.5, 3.5, 4.5, 5.5]) 

input_val_data = input_val_data - 3
output_val_data = output_val_data - 3

# 표로 출력
plt.plot(input_data, output_data, 'o')
plt.plot(input_val_data, output_val_data, 'o')
plt.show()
```

# 4. 모델 만들기

```python
# 4.모델 만들기
model = keras.models.Sequential()
model.add(keras.layers.Dense(units=16, input_shape=(1,),activation='relu'))
model.add(keras.layers.Dense(units=16, activation='relu'))
model.add(keras.layers.Dense(units=1,  activation=None))
```

# 5. 모델 학습과정 설정하기

```python
# 5.모델 학습과정 설정하기
model.compile(loss='mse', optimizer='sgd', metrics=['mse'])
```

# 6. 모델 학습시키기

```python
# 6.모델 학습시키기
history = model.fit(input_data, output_data, validation_data=(input_val_data, output_val_data), epochs=1000, batch_size=4)
```

# 7.결과 출력

```python
# Loss 출력
print(history.history.keys())
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(('train_loss', 'val_loss'))
plt.show()
```

```python
# 예측값 출력
yhat = model.predict(input_data)

plt.plot(input_data, output_data, 'o')
plt.plot(input_data, yhat)
plt.show()
```
